```python

import numpy as np
import keras
from keras.models import Sequential
from keras.layers.core import Dense
from keras.layers import Dropout,Conv2D
from sklearn.model_selection import train_test_split
import csv
import pandas as pd
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from sklearn.preprocessing import LabelBinarizer
encoder = LabelBinarizer()
main = pd.read_csv('dataset/nba.games.stats - nba_prev.csv')
home = pd.read_csv('dataset/nba.games.stats - nba_prev.csv',usecols=['Team'])
away = pd.read_csv('dataset/nba.games.stats - nba_prev.csv',usecols=['Opponent']).values
ha = pd.read_csv('dataset/nba.games.stats - nba_prev.csv',usecols=['Home']).values
pointsHome = pd.read_csv('dataset/nba.games.stats - nba_prev.csv',usecols=['TeamPoints']).values
pointsAway = pd.read_csv('dataset/nba.games.stats - nba_prev.csv',usecols=['OpponentPoints']).values
```

```python

home = encoder.fit_transform(home)
away = encoder.fit_transform(away)
homeaway = encoder.fit_transform(ha)
team = np.append(home,away,axis=1)
team = np.append(team,homeaway,axis=1)
```

```python
winLoss = pd.read_csv('dataset/nba.games.stats - nba_prev.csv',usecols=['WINorLOSS']).values
```

```python
winLoss = encoder.fit_transform(winLoss)
winLoss.shape
```

```python
points = np.append(pointsHome,pointsAway,axis=1)
points.shape
op = np.append(winLoss,points,axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(team, points, test_size=0.20)
```

```python
#MODEL
model = Sequential()
keras.optimizers.adam(lr=0.01)
model.add(Dense(50, activation='relu',input_dim=61))
model.add(Dropout(0.20))
model.add(Dense(46, activation='relu'))
# model.add(Dense(30, activation='sigmoid'))
model.add(Dense(25, activation='relu'))
model.add(Dense(10, activation='relu'))

# model.add(Dense(20, activation='relu'))
model.add(Dense(2, activation = 'relu'))

model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
history = model.fit(np.asarray(X_train),np.asarray(y_train), epochs=100)
```

```python
print(model.evaluate(X_test,y_test))
```
